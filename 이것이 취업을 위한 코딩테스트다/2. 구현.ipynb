```python
# 상하좌우
N = 5
commands = ["R", "R", "R", "U", "D", "D"]

def solution(N:int, commands: list[str]) -> list[int]:
    point = [1, 1]
    max = N
    command_x_table = {
        "R" : 1,
        "L" : -1
    }
    command_y_table = {
        "D" : 1,
        "U" : -1
    }

    for command in commands:
        if command in command_x_table:
            change_point = 1
            command_table = command_x_table
        else:
            change_point = 0
            command_table = command_y_table
        if not 0 < point[change_point] + command_table[command] <= max:
            continue
        point[change_point] += command_table[command]
    return point
solution(N, commands)

```

```python
# 시각
N = 5
def solution(N: int) -> int:
    count = 0
    for hour in range(N+1):
        for min in range(60):
            for sec in range(60):
                if '3' in str(hour) + str(min) + str(sec):
                    count += 1

    return count
solution(5)
```

```python
# 왕실의 나이트

input = 'c2'

def solution(input: str) -> int:
    cur_point = [ord(input[0])-ord('a')+1, int(input[1])]
    go_list = [(2, 1), (2, -1), (-2, 1), (-2, -1), (1, 2), (-1, 2), (1, -2), (-1, -2)]
    count = 0
    for go in go_list:
        if 0 < cur_point[0] + go[0] <= 8 and 0 < cur_point[1] + go[1] <= 8:
            count += 1
    return count

solution(input)
```

```python
# 게임 개발
N = 4
M = 4
A =1
B = 1
d = 0
maps = [[1,1,1,1],
       [1,0,0,1],
       [1,1,0,1],
       [1,1,1,1]]

def solution(N:int, M: int, A: int, B: int, d: int, maps: list[list[int]]) -> int:
    chack_map = [[0]*M for _ in range(N)]
    turn_time = 0
    chack_map[A][B] = 1

    dx = [-1,0,1,0]
    dy = [0,1,0,-1]

    while True:
        d = (d-1) % 4
        nx = A + dx[d]
        ny = B + dy[d]
        turn_time += 1

        if maps[nx][ny] == 0 and chack_map[nx][ny] == 0:
            chack_map[nx][ny] = 1
            A = nx
            B = ny
            turn_time = 0
        else:
            if turn_time == 4:
                turn_time = 0
                back_d = (d+2)%4 # 뒷 방향으로 가기
                nx = A + dx[back_d]
                ny = B + dy[back_d]
                if maps[nx][ny] == 1: # 뒤쪽이 바다라면 리턴
                    return sum(map(sum, chack_map))
                else: # 뒤쪽이 땅이라면 뒤쪽으로 이동
                    A = nx
                    B = ny



solution(N, M, A, B, d, maps)
```

```python
def solution(s):
    answer = len(s)

    # 1개 단위(step)부터 압축 단위를 늘려가며 확인
    for step in range(1, (len(s) // 2) +1):
        compressed = ""
        prev = s[0:step]
        count = 1
        #단위(step) 크기만큼 증가시키며 이전 문자열과 비교
        for j in range(step, len(s), step):
            #이전 상태와 동일하다면 압축 횟수(count) 증가
            if prev == s[j:j+step]:
                count += 1
            # 다른 문자열이 나왔다면 (더 이상 압축하지 못하는 경우라면)
            else:
                compressed += str(count) + prev if count >= 2 else prev
                prev = s[j:j + step] # 다시 상태 초기화
                count = 1
        # 남아 있는 문자열에 대해서 처리
        compressed += str(count) + prev if count >= 2 else prev
        # 만들어지는 압축 문자열이 가장 짧은 것이 정답
        answer = min(answer, len(compressed))
    return answer
solution("abcabcdede")
```

```python
#https://school.programmers.co.kr/learn/courses/30/lessons/60057/solution_groups?language=python3
def compress(text, tok_len):
    words = [text[i:i+tok_len] for i in range(0, len(text), tok_len)]
    res = []
    cur_word = words[0]
    cur_cnt = 1
    for a, b in zip(words, words[1:] + ['']):

        if a == b:
            cur_cnt += 1
        else:
            res.append([cur_word, cur_cnt])
            cur_word = b
            cur_cnt = 1
    return sum([len(word) + (len(str(cnt)) if cnt > 1 else 0) for word, cnt in res])

def solution(text):
    print(text)
    print( [len(text)])
    return min(compress(text, tok_len) for tok_len in list(range(1, int(len(text)/2) + 1)) + [len(text)])

a = [
    "aabbaccc",
    "ababcdcdababcdcd",
    "abcabcdede",
    "abcabcabcabcdededededede",
    "xababcdcdababcdcd",

    'aaaaaa',
]

for x in a:
    print(solution(x))
```

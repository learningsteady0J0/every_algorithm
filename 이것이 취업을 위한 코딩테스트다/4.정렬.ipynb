```python
array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]
def selection_sort(array: list[int]):
    for i in range(len(array)):
        min_index = i
        for j in range(i+1, len(array)):
            if array[j] < array[min_index]:
                min_index = j
        array[i], array[min_index] = array[min_index], array[i]

selection_sort(array)
print(array)
```

# 삽입 정렬
데이터를 하나씩 확인하며, 각 데이터를 적절한 위치에 삽입하기.
## 시간 복잡도 : O(N^2),  
## 최선의상황(모든 데이터가 정렬되어있을 경우 ) : O(N)

```python
array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]
def insertion_sort(array: list[int]):
    for i in range(1, len(array)):
        for j in range(i, 0, -1):
            if array[j] < array[j-1]:
                array[j], array[j-1] = array[j-1], array[j]
            else:
                break

insertion_sort(array)
print(array)









```

# 퀵 정렬
1. 기준 데이터를 설정하고 그 기준보다 큰 데이터와 작은 데이터의 위치를 바꾸기
2. 리스트를 반으로 나누기
숫자를 교환하기 위한 기준을 '피벗'이라고 한다.
리스트를 분할하는 방법에 따라 퀵 정렬이 구분되는데, 대표저긴 분할 방식인 호어(Hoare partition)분할 방식으로 구현.

1. 리스트에서 첫 번째 데이터를 피벗으로 정한다.

## 시간 복잡도 O(N log N )
## 최악의 상황: 이미 데이터가 정렬된 경우 O(N^2)

```python
array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]

def quick_sort_classic(array: list[int], start:int, end:int):
    if start >= end:
        return
    pivot = start
    left = start+1
    right = end

    while left <= right:
        while left <= end and array[pivot] > array[left]:
            left += 1
        while start < right and array[pivot] < array[right]:
            right -= 1
        if left > right:
            array[pivot], array[right] = array[right], array[pivot]
        else:
            array[left], array[right] = array[right], array[left]
    quick_sort_classic(array, start, right-1)
    quick_sort_classic(array, right+1, end)

quick_sort_classic(array, 0, len(array)-1)
print(array)

array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]

def quick_sort_python(array: list[int]) -> list[int]:
    if len(array) <= 1:
        return array

    pivot = array[0]
    tail = array[1:]

    left_side = [x for x in tail if x <= pivot]
    right_side = [x for x in tail if x > pivot]

    return quick_sort_python(left_side) + [pivot] + quick_sort_python(right_side)

print(quick_sort_python(array))
```

```python
array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]

def quick_sort_classic(array: list[int], start:int, end:int):
    if start >= end:
        return array
    pivot = start
    left = start + 1
    right = end

    while left <= right:
        while left <= end and array[left] <= array[pivot]:
            left += 1
        while start < right and array[pivot] <= array[right]:
            right -= 1
        if left > right:
            array[right], array[pivot] = array[pivot], array[right]
        else:
            array[left], array[right] = array[right], array[left]

    quick_sort_classic(array, start, right-1)
    quick_sort_classic(array, right+1, end)

quick_sort_classic(array, 0, len(array)-1)
print(array)

array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]

def quick_sort_python(array: list[int]) -> list[int]:
    if len(array) <= 1:
        return array

    pivot = array[0]
    tail = array[1:]

    left_side = [x for x in tail if x <= pivot]
    right_side = [x for x in tail if x > pivot]

    return quick_sort_python(left_side) + [pivot] + quick_sort_python(right_side)

print(quick_sort_python(array))
```

```python
array = [7, 5, 9, 0, 3, 1, 6, 2, 9, 1, 4, 8, 0, 5, 2,15, 20, 333]

def count_sort(array:list[int]):
    count = [0] * (max(array) + 1)

    for i in range(len(array)):
        count[array[i]] += 1

    for i in range(len(count)):
        for j in range(count[i]):
            print(i, end=' ')
count_sort(array)
```

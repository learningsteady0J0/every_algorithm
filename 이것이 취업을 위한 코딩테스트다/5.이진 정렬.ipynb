```python
# 부품 찾기
A = [8, 3, 7, 9, 2]
B = [5, 7, 9]

A.sort()

def binary(array, target, start, end):
    if start > end:
        return None

    mid = (start+end) // 2
    if array[mid] == target:
        return mid
    elif array[mid] > target:
        return binary(array, target, start, mid-1)
    elif array[mid] < target:
        return binary(array, target, mid+1, end)

for i in B:
    result = binary(A, i, 0, len(A)-1)
    if result == None:
        print("No", end=' ')
    else:
        print("Yes", end=' ')
```

```python
# 떡볶이 떡 만들기
M = 6
A = [19, 15, 10, 17]

def solution(A:list[int], M:int) -> int:
    start = 0
    end = max(A)
    result = 0

    while start <= end:
        mid = (start+end)//2
        total = 0

        for x in A:
            if x > mid:
                total += (x-mid)
        if total < M:
            end = mid-1
        else:
            result = mid
            start = mid+1
    return result


solution(A, M)
```

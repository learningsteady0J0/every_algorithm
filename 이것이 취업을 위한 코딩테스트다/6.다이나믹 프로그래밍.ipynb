```python
# 탑 다운 방식
d = [0] * 100

def pibo(x):
    if x == 1 or x == 2:
        return 1
    if d[x] != 0:
        return d[x]
    d[x] = pibo(x-1) + pibo(x-2)
    return d[x]
```

```python

def dfs(idx:int):
    global visited, answer
    visited[idx] = 1

    for i in graph[idx]:
        if visited[i] == 0:
            answer[i] = idx
            dfs(i)

N = int(input())
graph = [[] for _ in range(N+1)]
visited = [0] * (N+1)
answer = [0] * (N+1)

for _ in range(N):
    u, v = map(int, input().split())
    graph[u].append(v)
    graph[v].append(u)

dfs(1)

for i in range(2, N+1):
    print(answer[i])
```

```python
# 바텀 업 방식
d = [0] * 100

d[1] = 1
d[2] = 1
n = 99

for i in range(3, n+1):
    d[i] = d[i-1] + d[i-2]
print(d[n])
```

```python
# 1로 만들기
x = 199
d = [0] * 200
def solution(x:int):
    if x == 1:
        return 0
    if d[x] != 0:
        return d[x]

    d[x] = solution(x-1) + 1
    if x % 2 == 0:
        d[x] = min(d[x], solution(x//2)+1)
    elif x % 3 == 0:
        d[x] = min(d[x], solution(x//3)+1)
    elif x % 5 == 0:
        d[x] = min(d[x], solution(x//5)+1)
    return d[x]
solution(x)
print(d[x])

d = [0] * 200
def bottomup(x:int):
    for i in range(2, x+1):
        d[i] = d[i-1] +1
        if i % 2 == 0:
            d[i] = min(d[i], d[i//2] +1)
        elif i % 3 == 0:
            d[i] = min(d[i], d[i//3] +1)
        elif i % 5 == 0:
            d[i] = min(d[i], d[i//5] +1)

bottomup(x)
print(d[x])
```

```python
#개미 전사
N = 6
array = [4, 3, 1, 5,3,2]

d = [0] * (N+1)

d[0] = array[0]
d[1] = max(array[0], array[1])
for i in range(2,N):
    d[i] = max(d[i-1], d[i-2] + array[i])

print(d[N-1])
```

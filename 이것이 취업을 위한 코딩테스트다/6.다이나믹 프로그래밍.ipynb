```python
# 탑 다운 방식
d = [0] * 100

def pibo(x):
    if x == 1 or x == 2:
        return 1
    if d[x] != 0:
        return d[x]
    d[x] = pibo(x-1) + pibo(x-2)
    return d[x]
```

```python
# 1로 만들기
x = 199
d = [0] * 200
def solution(x:int):
    if x == 1:
        return 0
    if d[x] != 0:
        return d[x]

    d[x] = solution(x-1) + 1
    if x % 2 == 0:
        d[x] = min(d[x], solution(x//2)+1)
    elif x % 3 == 0:
        d[x] = min(d[x], solution(x//3)+1)
    elif x % 5 == 0:
        d[x] = min(d[x], solution(x//5)+1)
    return d[x]
solution(x)
print(d[x])

d = [0] * 200
def bottomup(x:int):
    for i in range(2, x+1):
        d[i] = d[i-1] +1
        if i % 2 == 0:
            d[i] = min(d[i], d[i//2] +1)
        elif i % 3 == 0:
            d[i] = min(d[i], d[i//3] +1)
        elif i % 5 == 0:
            d[i] = min(d[i], d[i//5] +1)

bottomup(x)
print(d[x])
```

```python
#개미 전사
N = 6
array = [4, 3, 1, 5,3,2]

d = [0] * (N+1)

d[0] = array[0]
d[1] = max(array[0], array[1])
for i in range(2,N):
    d[i] = max(d[i-1], d[i-2] + array[i])

print(d[N-1])
```

```python
#바닥 공사
N = 4
dp = [0] * (N+1)

dp[1] = 1
dp[2] = 3
for i in range(3, N+1):
    dp[i] = (dp[i-2]*2) + dp[i-1]
print(dp)
```

```python
#효율적인 화폐 구성

n = 4
m = 15
array = [3, 5, 2, 7]
dp = [10000] * (m+1)

dp[0] = 0
for i in range(n):
    for j in range(array[i], m+1):
        dp[j] = min(dp[j], dp[j-array[i]] + 1)

print(dp)

```

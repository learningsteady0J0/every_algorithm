```python
# 큰 수의 법칙
# 다양한 수로 이루어진 배열이 있을 때 주어진 수들을 M번 더하여 가장 큰 수를 만들자
# 단, 배열의 특정한 인덱스에 해당하는 수가 연속 K번을 초과해서 더해질 수 없다.
# in : 배열의 크기(N): 5, 더해지는 횟수(M): 8, 중첩횟수(K): 3 , [2, 4, 5, 4, 6]
# out : 46

n = 5
m = 8
k = 3
num_list = [2, 4, 5, 4, 6]

def solution(n: int, m: int, k: int, num_list: list[int] ) -> int:
    num_list.sort(reverse=True)
    print(num_list)
    sum = 0
    k_count = 0

    for _ in range(m):
        if k_count != k:
            sum += num_list[0]
        else:
            sum += num_list[1]
            k_count = 0
            continue
        k_count += 1
    return sum

def solution(n: int, m: int, k: int, num_list: list[int] ) -> int:
    num_list.sort()
    first = num_list[-1]
    second = num_list[-2]

    count = m // (k+1) * k
    count += m % (k+1)

    sum = 0
    sum += count * first
    sum += (m-count) * second

    return sum

solution(n, m, k, num_list)
```

```python
import sys
sys.setrecursionlimit(10**6)


M, N, H = map(int, input().split())
map_ = [[[0] * (M+2) for _ in range(N+2)] for _ in range(H + 2)]
visited = [[[0] * (M+2) for _ in range(N+2)] for _ in range(H + 2)]

for h in range(H):
    for y in range(N):
        rows = list(map(int, input().split()))
        for x in range(M):
            map_[h+1][y+1][x+1] = rows[x]

# 모든 토마토가 익어 있음
all_check = True
for h in range(1, H+1):
    for y in range(1, N+1):
        if 0 in map_[h][y][1:-1]:
            all_check = False

# 토마토가 모두 익지 못함
dx = [0,0,0,0,-1,1]
dy = [0,0,1,-1,0,0]
dh = [1,-1,0,0,0,0]
not_check = False
for h in range(1, H+1):
    for y in range(1, N+1):
        for x in range(1, M+1):
            count = 0
            if map_[h][y][x] == 0:
                for i in range(6):
                    new_x = x + dx[i]
                    new_y = y + dy[i]
                    new_h = h + dh[i]
                    if (1 <= new_x <= M) and (1 <= new_y <= N) and (1 <= new_h <= H):
                        if map_[new_h][new_y][new_x] == -1:
                            count += 1
            if count == 6:
                not_check = True


def bfs(h, y, x):
    global visited
    visited[h][y][x] = 1

    for i in range(6):
        new_x = x + dx[i]
        new_y = y + dy[i]
        new_h = h + dh[i]
        if (1 <= new_x <= M) and (1 <= new_y <= N) and (1 <= new_h <= H):
            if map_[new_h][new_y][new_x] == 0 and not visited[new_h][new_y][new_x]:
                map_[new_h][new_y][new_x] = 1
                visited[new_h][new_y][new_x] = 1

# 토마토가 익는 최소 날짜
count = 0
while True:
    for h in range(1, H+1):
        for y in range(1, N+1):
            for x in range(1, M+1):
                if (map_[h][y][x] == 1) and (not visited[h][y][x]):
                    bfs(h, y, x)
    count += 1
    check = True
    for h in range(1, H+1):
        for y in range(1, N+1):
            if 0 in map_[h][y][1:-1]:
                check = False
    if check:
        break

if all_check:
    print(0)
elif not_check:
    print(-1)
else:
    print(count)
```

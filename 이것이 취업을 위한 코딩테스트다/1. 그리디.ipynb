```python
# 큰 수의 법칙
# 다양한 수로 이루어진 배열이 있을 때 주어진 수들을 M번 더하여 가장 큰 수를 만들자
# 단, 배열의 특정한 인덱스에 해당하는 수가 연속 K번을 초과해서 더해질 수 없다.
# in : 배열의 크기(N): 5, 더해지는 횟수(M): 8, 중첩횟수(K): 3 , [2, 4, 5, 4, 6]
# out : 46

n = 5
m = 8
k = 3
num_list = [2, 4, 5, 4, 6]

def solution(n: int, m: int, k: int, num_list: list[int] ) -> int:
    num_list.sort(reverse=True)
    print(num_list)
    sum = 0
    k_count = 0

    for _ in range(m):
        if k_count != k:
            sum += num_list[0]
        else:
            sum += num_list[1]
            k_count = 0
            continue
        k_count += 1
    return sum

def solution(n: int, m: int, k: int, num_list: list[int] ) -> int:
    num_list.sort()
    first = num_list[-1]
    second = num_list[-2]

    count = m // (k+1) * k
    count += m % (k+1)

    sum = 0
    sum += count * first
    sum += (m-count) * second

    return sum

solution(n, m, k, num_list)
```

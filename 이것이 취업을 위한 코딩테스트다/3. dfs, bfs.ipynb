```python
# 음료수 얼려 먹기
N = 15
M = 14
ice = ["00000111100000",
       "11111101111110",
       "11011101101110",
       "11011101100000",
       "11011111111111",
       "11011111111100",
       "11000000011111",
       "01111111111111",
       "00000000011111",
       "01111111111000",
       "00011111111000",
       "00000001111000",
       "11111111110011",
       "11100011111111",
       "11100011111111"]

'''
N = 4
M = 5
ice = ["00110",
       "00011",
       "11111",
       "00000"]
'''

import sys
sys.setrecursionlimit(10**6)


dx = [0,0,-1,1]
dy = [-1,1,0,0]
def dfs(y:int, x:int):
    global visited
    visited[y][x] = 1

    for i in range(4):
        new_x = x+dx[i]
        new_y = y+dy[i]
        if map_[new_y][new_x] and not visited[new_y][new_x]:
            dfs(new_y,new_x)



map_ = [[0] * (M+2) for _ in range(N+2)]
visited = [[0] * (M+2) for _ in range(N+2)]

for i in range(N):
    rows = ice[i]
    for j in range(M):
        if int(rows[j]) == 1:
            map_[i+1][j+1] = 0
        else:
            map_[i+1][j+1] = 1

count = 0
for y in range(1, N+1):
    for x in range(1, M+1):
        if map_[y][x] and not visited[y][x]:
            dfs(y, x)
            count+=1
print(count)
```

```python
# 미로 탈출
N, M = 5, 6
map = ["101010",
       "111111",
       "000001",
       "111111",
       "111111"]

map_ = [[0] * (M+2) for _ in range(N+2)]
visited = [[0] * (M+2) for _ in range(N+2)]

for i in range(N):
    rows = map[i]
    for j in range(M):
        map_[i+1][j+1] = int(rows[j])

import collections
queue = collections.deque()
queue.append((1,1,1))

dx = [1,-1,0,0]
dy = [0,0,1,-1]

while queue:
    y,x,count = queue.popleft()
    if y == N and x == M:
        print(count)
        break
    if not visited[y][x]:
        visited[y][x] = 1
        for i in range(4):
            new_x = x + dx[i]
            new_y = y + dy[i]
            if map_[new_y][new_x] and not visited[new_y][new_x]:
                queue.append((new_y,new_x,count+1))

```

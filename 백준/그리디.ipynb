1. 콘센트에 꼽혀 있는 전자기기의 다음 선택까지의 거리를 기준으로 빼면 어떨까?
2. hash table에 각 전자기기의 다음 idx를 기입한다.
3. 콘센트를 빼야하는 상황이 오면 hash table에 접근해서 가장 먼 거리를 가진 전자기기를 빼낸다.

한번의 for문에서 각각의 전자기기의 idx를 기록하자.
그리고 각 전자기기의 현재 idx를 기록하면 다음 idx까지의 거리를 구할 수 있을 것 같다.

더 진행하다보니, 현재 array idx와 콘센트에 꼽혀있는 전자기기의 다음 idx와의 거리를 비교하면 될 거 같다.

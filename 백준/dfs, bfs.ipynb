# 백준 2573번 - 빙산
### 답안 - https://www.acmicpc.net/submit/2573/70262324
### 생각의 흐름 -
- goal : 빙산이 2 덩어리가 되는 최초의 시간 구하기
- 필요한 모듈 : 빙산이 녹는 모듈, 빙산 덩어리를 계산하는 모듈
- 빙산 덩어리를 계산하는 모듈 -> 전체 map_을 체크해서 1이상의 높이를 가지고 아직 방문하지 않은 빙산들을 체크하면된다.
- 빙산이 녹는 모듈
1. 전체 map_을 체크해서 1이상의 높이를 가지는 첫번째 빙산조각의 인덱스로 dfs를 시작한다.
2. 해당 인덱스에서 상하좌우를 체크하여 0 이 있다면 그 개수만큼 줄어든다.
3. 이번턴에 0이 되어있는거에 대해서는 카운트하면 안되니깐, visited를 활용해서 방문하지 않은 0만 체킹한다.

- python으로는 시간초과가 뜸. -> pypy3로 바꾸었는데 메모리초과가 뜸. -> pypy3는 재귀에 약해서 재귀의 최대범위를 작게해줘야한다. -> 기존의 10**6을 10**4로 바꾸니 메모리초과가 뜨지않고 정답처리됨.

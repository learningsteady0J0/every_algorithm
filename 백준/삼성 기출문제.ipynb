- 필요한 모듈
    1. 파이어볼의 이동
        - 1.1 각좌표를 돌아다니면서 list의 원소가 존재하면 모두 실행 (파이어볼이 2개 이상일 수 있으니)

    3. 남아 있는 파이어볼 질량의 합 구하기

- 파이어볼의 좌표를 순환시켜야 했는데 난잡하게 구현한 것 같다. 조금 더 깔끔하게 구현할 수 없을까?

    ```python
    ny = (y + dy[d]*s) % N
    nx = (x+ dx[d]*s) % N
    if ny == 0: ny = N
    if nx == 0: nx = N
    ```

    - ```python
    map_ = [[list() for _ in range(N+2) ]  for _ in range(N+2)]
    ```

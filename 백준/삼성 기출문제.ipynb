- 도시의 치킨 거리 : 모든 집의 치킨 거리의 합

# 백준 3190 - 뱀
### 걸린 시간 : 47분
### https://www.acmicpc.net/submit/3190/70590427
### 생각의 흐름
- goal : dummy게임이 몇 초에 끝나는지 출력

    1. 뱀의 머리가 다음칸에 위치한다.
    2. 만약 벽이나 자기자신의 몸과 부딪히면 게임이 끝난다
    3. 이동한 칸에 사과가 있으면, 그 칸의 사과가 없어지고 꼬리가 안움직인다.
    4. 이동한 칸에 사과가 없으면, 꼬리가 위치한 칸을 비워준다.

- 이동 : 꼬리의 좌표를 0으로 가려고하는 방향의 좌표를 1로 바꿔주기 단, 사과가 있으면 머리 좌표만 1.
- snake : deque를 이용해서 전체 좌표를 관리해주자. 머리를 추가하고 꼬리를 없앨때 간편한다.

# 백준 14499 - 주사위 굴리기
### 걸린 시간 : 1시간 15분
### https://www.acmicpc.net/submit/14499/70633658
### 생각의 흐름
- goal : 이동할 때마다 주사위의 윗 면에 쓰여 있는 수를 출력한다. 단, 바깥으로 이동하려고하면 명령을 무시하고 출력하지 않는다.

- 필요한 모듈
    1. 주사위 자료구조
        - 1.1 전개도를 기준으로 생각해보자.
    2. 주사위를 굴리는 모듈
        - 2.1 동쪽 1, 서쪽 2, 북쪽 3, 남쪽 4
    3. 지도와 주사위가 맞닿는면을 보고 처리하는 모듈
        - 3.1 이동한 칸에 쓰여 있는 수가 0 : 주사위 박다면에 쓰여있는수가 칸에 복사.
        - 3.2 0이 아니면 칸에 쓰여있는 수가 주사위의 바닥면에 복사. 칸은 0이 됨

- 위로 굴렸을 때 (A : 상하전개도, B:좌우전개도)
    - A: 2 1 5 6 -> 6 2 1 5 -> 5 6 2 1 -> 1 5 6 2
    - B: 4 1 3 6 -> 4 2 3 5 -> 4 6 3 1 -> 4 5 3 2
- 옆으로 굴렸을 때
    - A: 2 1 5 6 -> 2 4 5 3 -> 2 6 5 1
    - B: 4 1 3 6 -> 6 4 1 3 -> 3 6 4 1
- 밑면 : 어떤 리스트든 마지막 [3], 윗면 : 어떤리스트든 두번째[1]
- 북쪽으로 굴릴 땐 A를 rotate(1)하고 B는 이전[0], A[1], 이전[2], A[3]
- 남쪽으로 굴릴 땐 A를 rotate(-1)하고 B는 이전[0], A[1], 이전[2], A[3]
- 동쪽으로 굴릴 땐 B를 rotate(1)하고 A는 이전[0], B[1], 이전[2], B[3]
- 서쪽으로 굴릴 땐 B를 rotate(-1)하고 A는 이전[0], B[1], 이전[2], B[3]
------
- 이 문제에선 x가 0~N-1로써 행에 해당한다. 문제를 잘 읽자.

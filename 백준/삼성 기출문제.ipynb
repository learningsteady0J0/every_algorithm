- goal : 이동할 때마다 주사위의 윗 면에 쓰여 있는 수를 출력한다. 단, 바깥으로 이동하려고하면 명령을 무시하고 출력하지 않는다.

- 필요한 모듈
    1. 주사위 자료구조
        - 1.1 전개도를 기준으로 생각해보자.
    2. 주사위를 굴리는 모듈
        - 2.1 동쪽 1, 서쪽 2, 북쪽 3, 남쪽 4
    3. 지도와 주사위가 맞닿는면을 보고 처리하는 모듈

        - 3.2 0이 아니면 칸에 쓰여있는 수가 주사위의 바닥면에 복사. 칸은 0이 됨

    - A: 2 1 5 6 -> 2 4 5 3 -> 2 6 5 1
    - B: 4 1 3 6 -> 6 4 1 3 -> 3 6 4 1

- 남쪽으로 굴릴 땐 A를 rotate(-1)하고 B는 이전[0], A[1], 이전[2], A[3]
- 동쪽으로 굴릴 땐 B를 rotate(1)하고 A는 이전[0], B[1], 이전[2], B[3]
- 서쪽으로 굴릴 땐 B를 rotate(-1)하고 A는 이전[0], B[1], 이전[2], B[3]

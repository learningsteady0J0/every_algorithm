# 백준 15686 - 치킨 배달
### 걸린 시간 : 45분
### https://www.acmicpc.net/submit/15686/70524600
### 생각의 흐름
- goal : 도시에 있는 치킨집 중에서 최대 M개를 고르고, 도시의 치킨 거리가 가장 작은 값 출력
- 치킨 거리 : 집과 가장 가까운 치킨집 사이의 거리.
- 도시의 치킨 거리 : 모든 집의 치킨 거리의 합

- 필요한 모듈
    - 치킨 거리를 구하는 모듈.
    - 모든 치킨 집의 경우의 수 구하기 -> 백트래킹
    - 그 중에서 가장 작은 도시의 치킨 거리 출력

- 치킨 거리를 구하는 모듈.
    - 치킨집과 집의 좌표는 미리 정해져 있다.
    - dist=[{ 가게 1: 거리, 가게 2: 거리 .. }] 형태로 각 집에서 각 가게와의 거리를 미리 구해놓자.
- 백트래킹
    - 치킨집을 선택할 수 있는 모든 경우의 수를 백트래킹으로 접근하고, 선택된 idx를 list형태로 치킨의 거리를 구하는 함수에 주게 되면 위에서 저장된 dist를 이용해서 도시의 치킨 거리를 쉽게 구할 수 있다.

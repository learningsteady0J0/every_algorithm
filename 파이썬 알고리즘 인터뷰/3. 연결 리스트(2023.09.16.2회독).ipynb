```python
# 홀짝 연결 리스트
# 연결 리스트를 홀수 노드 다음에 짝수 노드가 오도록 재구성하라. 공간 O(1),시간 O(n)에 풀이하라.
# in : 1 -> 2 -> 3 -> 4 -> 5 -> NULL
# out : 1 -> 3 -> 5 -> 2 -> 4 -> NULL
# in : 2 -> 1 -> 3 -> 5 -> 6 -> 4 -> 7 -> NULL
# out : 2 -> 3 -> 6 -> 7 -> 1 -> 5 -> 4 -> NULL
head = ListNode(2)
head.next = ListNode(1)
head.next.next = ListNode(3)
head.next.next.next = ListNode(5)
head.next.next.next.next = ListNode(6)
head.next.next.next.next.next = ListNode(4)
head.next.next.next.next.next.next = ListNode(7)

def odd_even_list(head: ListNode) -> ListNode:
    #예외 처리
    if head is None:
        return None

    odd = head
    even, even_head = head.next, head.next

    while even and even.next:
        odd.next, even.next = odd.next.next, even.next.next
        odd, even = odd.next, even.next

    odd.next = even_head
    return head

head = odd_even_list(head)
while head:
    print(head.val)
    head = head.next
```

```python
# 역순 연결 리스트2
# 인덱스 m에서 n까지를 역순으로 만들어라. 인덱스 m은 1부터 시작한다.
# in : 1 -> 2 -> 3 -> 4 -> 5 -> NULL, m = 2, n = 4
# out : 1 -> 4 -> 3 -> 2 -> 5 -> NULL
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)

def reverse_list(head:ListNode, m:int, n:int) -> ListNode:
    start = root = ListNode(None)
    start.next = head
    for _ in range(m-1):
        start = start.next
    end = start.next

    for _ in range(n-m):
        tmp, start.next, end.next = start.next, end.next, end.next.next
        start.next.next = tmp
    return root.next


head = reverse_list(head, 1, 4)
while head:
    print(head.val)
    head = head.next
```
